```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_eco = pd.read_csv("../data/contrywise_economicdata.csv")
data_country = pd.read_csv("../data/country_wise_latest.csv")
```

```python
data = pd.merge(data_country,data_eco,left_on="Country/Region",right_on="Country")
```

```python
data
```

```python
data.columns
```

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(fig, left=True, bottom=True)
sns.scatterplot(x="Deaths / 100 Cases",y="Recovered / 100 Cases",hue="WHO Region",data=data,
                sizes=(1, 8), linewidth=0, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(fig, left=True, bottom=True)
data["Population","New cases", 'GDP', 'GDP(PPP)','Inflation Rate']
sns.scatterplot(x="Deaths / 100 Cases",y="Recovered / 100 Cases",hue="WHO Region",data=data,
                sizes=(1, 8), linewidth=0, ax=ax)
```
